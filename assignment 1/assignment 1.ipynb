```python
import requests
import re
from bs4 import BeautifulSoup
import csv
import codecs
import pandas as pd
r = requests.get('https://archiveofourown.org/media')
html_str = r.text #get the content of the request
data = BeautifulSoup(html_str,"html.parser")
    
book_session = data.find_all('li', attrs={'class':"medium listbox group"})
books=[]
for session in book_session:
    elements = session.find_all('li')
    for element in elements:
        book={}
        book['genre']=((element.find('a').parent.parent.parent)).find('h3',attrs={'class':"heading"}).text
        book['name'] = element.find('a').text
        book['reviews'] = element.text.strip().split('\n')[-1].strip()
        books.append(book)

fileName = 'PythonBookreviews.csv'

with codecs.open(fileName, 'w', 'utf-8') as csvfile:
#指定 csv 文件的头部显示项
    filednames = ['genre','name', 'reviews']
    writer = csv.DictWriter(csvfile, fieldnames=filednames)
    writer.writeheader()
    for book in books:
        try:
            writer.writerow({'genre':book['genre'],'name':book['name'], 'reviews':book['reviews']})
        except UnicodeEncodeError:
            print("编码错误, 该数据无法写到文件中, 直接忽略该数据")
```

```python
import requests
import re
from bs4 import BeautifulSoup
import csv
import codecs

import xlwt
import xlsxwriter

    
r = requests.get('https://archiveofourown.org/media')
html_str = r.text #get the content of the request
data = BeautifulSoup(html_str,"html.parser")
   
book_session = data.find_all('li', attrs={'class':"medium listbox group"})
books=[]

for session in book_session:
    elements = session.find_all('li')
    for element in elements:
        book={}
        
        book['genre']=((element.find('a').parent.parent.parent)).find('h3',attrs={'class':"heading"}).text
        book['name'] = element.find('a').text
        book['reviews'] = element.text.strip().split('\n')[-1].strip()
        books.append(book)

        workbook=xlsxwriter.Workbook('data1.xlsx')
        worksheet=workbook.add_worksheet('data')
        worksheet_head=['genre','name','reviews']

        for h in range(len(worksheet_head)):
               worksheet.write(0,h,worksheet_head[h])
    row = 1
    col = 0
for content in books:
    worksheet.write(row, col, content['genre'])
    worksheet.write(row, col + 1, content['name'])
    worksheet.write(row, col + 2, content['reviews'])
    row += 1


workbook.close()
```
